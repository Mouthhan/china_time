```python
#讀0蓋1
import requests
from bs4 import BeautifulSoup 
import urllib.request
import pandas as pd
import os
import datetime
import re
def get_chinatime_news(i,Title,Type,Text):
    #url i為第幾頁
    chinatime_url='https://www.chinatimes.com/realtimenews/?page='
    chinatime_url=chinatime_url+str(i)+'&chdtv'
    #創建陣列
    news_list = []
#     type_list = []
#     title_list = []
#     content_list=[]
    headers = {'User-Agent':'Mozilla/5.0 (Windows NT 6.1; WOW64; rv:23.0) Gecko/20100101 Firefox/23.0'}
    req = urllib.request.Request(url=chinatime_url, headers=headers)
    html=urllib.request.urlopen(req).read()
    html = html.decode("utf-8") #解碼為unicode碼
    soup = BeautifulSoup(html, 'html.parser')
    h3_tag=soup.find_all("h3",class_="title")
    div_tag = soup.find_all("div", class_="category")
    for h3 in h3_tag:
        put=str(h3)
        Title.append(h3.string)#新聞標題
        visit_url='https://www.chinatimes.com'
        visit_url=visit_url+put[27:62]+'?chdtv'
        news_list.append(visit_url)#新聞連結
    for div in div_tag:
        Type.append(div.get_text())
    for k in range(0,len(news_list)):
        news_url=news_list[k]#要存取哪則新聞
        request = urllib.request.Request(url=news_url, headers=headers)
        content=urllib.request.urlopen(request).read()#新聞的HTML
        content = content.decode("utf-8") #解碼為unicode碼
        out = BeautifulSoup(content, 'html.parser')#解碼後
        text=out.find("meta",itemprop="description")#內文
        T=str(text)#字串
        end=len(T)-52
        result_text=''
        for n in range(15,end):
            result_text=result_text+(T[n])
        Text.append(result_text)
#     df=pd.DataFrame(list(zip(type_list,title_list,content_list)),columns=['類型','標題','內文'])
#     df.to_csv('ChinaTime_news2.csv',index=False,encoding='utf_8_sig')

        

    
if __name__ == '__main__':
    data=pd.read_csv('ChinaTime_data.csv')
    data_df=pd.DataFrame(data)
    Title=[]
    Type=[]
    Text=[]
    for title in data_df['Title']:
        Title.append(title)
    for typ in data_df['Type']:
        Type.append(typ)
    for text in data_df['Text']:
        Text.append(text)
    Min=len(Type)
    for i in range(1,11):
        get_chinatime_news(i,Title,Type,Text)
    df=pd.DataFrame(list(zip(Type,Title,Text)),columns=['Type','Title','Text'])
    for count in range(Min,Min+200):
        if df.iloc[count,0] == "政治" or df.iloc[count,0] =="兩岸":
            df.iloc[count,0]=0
        elif df.iloc[count,0] == "生活":
            df.iloc[count,0]=1
        elif df.iloc[count,0] == "國際":
            df.iloc[count,0]=2
        elif df.iloc[count,0] == "體育":
            df.iloc[count,0]=3
        elif df.iloc[count,0] == "娛樂":
            df.iloc[count,0]=4
        elif df.iloc[count,0] == "社會":
            df.iloc[count,0]=5
        elif df.iloc[count,0] == "財經":
            df.iloc[count,0]=6
        else:
            df.iloc[count,0]=7
    df.drop([7])#丟掉不符合ㄉ
    df_out=df[df.Type != 7]
    df_out.to_csv('ChinaTime_data1.csv',index=False,encoding='utf_8_sig')#輸出的拉
```
